```python
avg_co2_over_time = df.groupby('period')['value'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(x='period', y='value', data=avg_co2_over_time)
plt.title('Average CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('Average CO2 Emission (million metric tons)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg_co2_by_sector = df.groupby('sector-name')['value'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.barplot(x='sector-name', y='value', data=avg_co2_by_sector, palette='viridis')
plt.title('Average CO2 Emissions by Sector')
plt.xlabel('Sector')
plt.ylabel('Average CO2 Emission (million metric tons)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
